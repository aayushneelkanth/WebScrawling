```python
import requests
import urllib.request
```

```python
website_scrawl="http://midas.iiitd.edu.in/"
```

```python
data=urllib.request.urlopen(website_scrawl)
```

```python
from bs4 import BeautifulSoup
```

```python
datascrawl=BeautifulSoup(data)
```

```python
print(datascrawl)
```

```python
para_data=datascrawl.find_all('p')
for para in para_data:
    print(para.text)
```

```python
f=open('datainpara.txt','w')
for para in para_data:
    f.write(para.get_text("")+'\n')
f.close()
```

```python
f=open("datainpara.txt",'r')
print (f.read())
```
